```python
import math
from datetime import datetime,timedelta
import numpy as np
import pandas
import random
import preprocess_functions
```

```python
df1 = pandas.read_csv("training_20min_avg_volume_etc_cartype.csv")
df2 = pandas.read_csv("training2_20min_avg_volume_etc_cartype.csv")
df = df1.append(df2, ignore_index=True)
df_time_temp = df['time_window'].str.split('[',expand=True)[1].str.split(',',expand=True)[0]
df['time_start'] = df_time_temp
```

```python
df_add_time = preprocess_functions.add_time_dimension(df, df_time_temp)
```

```python
df_truthm = df_add_time[(df_add_time['month']==10) & (df_add_time['date']>17) 
            & (df_add_time['time']>23) & (df_add_time['time']<30)].reset_index().drop(['index'],axis=1)
df_trutha = df_add_time[(df_add_time['month']==10) & (df_add_time['date']>17) 
            & (df_add_time['time']>50) & (df_add_time['time']<57)].reset_index().drop(['index'],axis=1)
dm_s = df_truthm.sort_values(['tollgate_id','direction','time']).reset_index().drop(['index'],axis=1)
da_s = df_trutha.sort_values(['tollgate_id','direction','time']).reset_index().drop(['index'],axis=1)
df_truth = dm_s.append(da_s,ignore_index=True)
df_truth
```

```python
df_1_0 = preprocess_functions.extract_tg_dir(df, 1, 0)
df_1_1 = preprocess_functions.extract_tg_dir(df, 1, 1)
df_2_0 = preprocess_functions.extract_tg_dir(df, 2, 0)
df_3_0 = preprocess_functions.extract_tg_dir(df, 3, 0)
df_3_1 = preprocess_functions.extract_tg_dir(df, 3, 1)
df_1_0
```

```python
df_1_0_vacancy = df_1_0[df_1_0['volume']==0]
df_1_1_vacancy = df_1_1[df_1_1['volume']==0]
df_2_0_vacancy = df_2_0[df_2_0['volume']==0]
df_3_0_vacancy = df_3_0[df_3_0['volume']==0]
df_3_1_vacancy = df_3_1[df_3_1['volume']==0]
frame_fill = [df_1_0_vacancy,df_1_1_vacancy,df_2_0_vacancy,df_3_0_vacancy,df_3_1_vacancy]
df_vacancy = pandas.concat(frame_fill,ignore_index=True)
df_vacancy_new = df_vacancy['time_start'].apply(str)
df_vacancy.to_csv('phase2_wait_fill_vacancy.csv',index=False)
```

```python
df_weather1 = pandas.read_csv("v3_weather (table 7)_training.csv")
df_weather2 = pandas.read_csv("v2_weather (table 7)_test.csv")
df_weather_temp = df_weather1.append(df_weather2, ignore_index=True)
```

```python
df_waitfill = pandas.read_csv('phase2_wait_fill_vacancy.csv')
df_waitfill_temp = df_waitfill['time_start']
df_waitfill_insert_weather = preprocess_functions.insert_weather(df_waitfill, df_waitfill_temp, df_weather_temp)
df_waitfill_add_time = preprocess_functions.add_time_dimension(df_waitfill_insert_weather, df_waitfill_temp)
df_final_test = preprocess_functions.holiday_judge(df_waitfill_add_time, df_waitfill_temp)
df_final_test
```

```python
df_addition = pandas.DataFrame({'time':[i for i in xrange(72)]})
df_final_test = df_final_test.append(df_addition,ignore_index=True)
df_final_test = df_final_test.fillna('2016-3-19 00:00:00')
df_final_test['time_hour'] = df_final_test['time_start'].str.split('-',expand=True)[2]\
                             .str.split(' ',expand=True)[1].str.split(':',expand=True)[0].apply(int)
df_final_test['date'] = df_final_test['time_start'].str.split('-',expand=True)[2]\
                             .str.split(' ',expand=True)[0].apply(int)
df_final_test['minute'] = df_final_test['time_start'].str.split('-',expand=True)[2]\
                             .str.split(' ',expand=True)[1].str.split(':',expand=True)[1].apply(int)
```

```python
preprocess_functions.insert_onehot(df_final_test, 'tollgate_id')
preprocess_functions.insert_onehot(df_final_test, 'month')
preprocess_functions.insert_onehot(df_final_test, 'weekday')
preprocess_functions.insert_onehot(df_final_test, 'time')
preprocess_functions.insert_onehot(df_final_test, 'precipitation')
df_final_test
```
