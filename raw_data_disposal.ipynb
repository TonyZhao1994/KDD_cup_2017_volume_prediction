```python
import math
from datetime import datetime,timedelta
import numpy
```

```python
# %load E:\personal materials\KDD Traffic Flow Prediction\Scripts\scripts\aggregate_volume
#!/usr/bin/env python

"""
Calculate volume for each 20-minute time window.
"""

file_suffix = '.csv'
path = 'C:/Users/Tony/Jupyter_File/volume_prediction/dataSet_phase2/'  # set the data directory

in_file = 'volume(table 6)_test2'
out_suffix = '_20min_avg_volume_etc_cartype'
in_file_name = in_file + file_suffix
out_file_name = in_file.split('_')[1] + out_suffix + file_suffix

# Step 1: Load volume data
fr = open(path + in_file_name, 'r')
fr.readline()  # skip the header
vol_data = fr.readlines()
fr.close()

# Step 2: Create a dictionary to caculate and store volume per time window
volumes = {}  # key: time window value: dictionary
volume_model = {}
volume_type = {}
volume_has_etc = {}
for i in range(len(vol_data)):
    each_pass = vol_data[i].replace('"', '').split(',')
    tollgate_id = each_pass[1]
    direction = each_pass[2]
    car_model = each_pass[3]
    has_etc = each_pass[4]
    car_type = each_pass[5]
    pass_time = each_pass[0]

    pass_time = datetime.strptime(pass_time, "%Y-%m-%d %H:%M:%S")
    time_window_minute = int(math.floor(pass_time.minute / 20) * 20)
    #print pass_time
    start_time_window = datetime(pass_time.year, pass_time.month, pass_time.day,
                                 pass_time.hour, time_window_minute, 0)

    if start_time_window not in volumes:
        volumes[start_time_window] = {}
    if tollgate_id not in volumes[start_time_window]:
        volumes[start_time_window][tollgate_id] = {}
    if direction not in volumes[start_time_window][tollgate_id]:
        volumes[start_time_window][tollgate_id][direction] = 1
    else:
        volumes[start_time_window][tollgate_id][direction] += 1

    if start_time_window not in volume_model:
        volume_model[start_time_window] = {}
    if tollgate_id not in volume_model[start_time_window]:
        volume_model[start_time_window][tollgate_id] = {}
    if direction not in volume_model[start_time_window][tollgate_id]:
        volume_model[start_time_window][tollgate_id][direction] = {}
    if car_model not in volume_model[start_time_window][tollgate_id][direction]:
        volume_model[start_time_window][tollgate_id][direction][car_model] = 1
    else:
        volume_model[start_time_window][tollgate_id][direction][car_model] += 1

    if start_time_window not in volume_type:
        volume_type[start_time_window] = {}
    if tollgate_id not in volume_type[start_time_window]:
        volume_type[start_time_window][tollgate_id] = {}
    if direction not in volume_type[start_time_window][tollgate_id]:
        volume_type[start_time_window][tollgate_id][direction] = {}
    if car_type not in volume_type[start_time_window][tollgate_id][direction]:
        volume_type[start_time_window][tollgate_id][direction][car_type] = 1
    else:
        volume_type[start_time_window][tollgate_id][direction][car_type] += 1

    if start_time_window not in volume_has_etc:
        volume_has_etc[start_time_window] = {}
    if tollgate_id not in volume_has_etc[start_time_window]:
        volume_has_etc[start_time_window][tollgate_id] = {}
    if direction not in volume_has_etc[start_time_window][tollgate_id]:
        volume_has_etc[start_time_window][tollgate_id][direction] = {}
    if has_etc not in volume_has_etc[start_time_window][tollgate_id][direction]:
        volume_has_etc[start_time_window][tollgate_id][direction][has_etc] = 1
    else:
        volume_has_etc[start_time_window][tollgate_id][direction][has_etc] += 1

# Step 3: format output for tollgate and direction per time window
fw = open(out_file_name, 'w')
fw.writelines(','.join(['"tollgate_id"', '"time_window"', '"direction"', '"volume"', '"model0"',\
                       '"model1"', '"model2"', '"model3"', '"model4"', '"model5"', '"model6"',\
                       '"model7"', '"type0"', '"type1"', '"has_etc1"']) + '\n')
time_windows = list(volumes.keys())
time_windows.sort()
for time_window_start in time_windows:
    time_window_end = time_window_start + timedelta(minutes=20)
    for tollgate_id in volumes[time_window_start]:
        for direction in volumes[time_window_start][tollgate_id]:
            if '0' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['0'] = 0
            if '1' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['1'] = 0
            if '2' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['2'] = 0
            if '3' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['3'] = 0
            if '4' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['4'] = 0
            if '5' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['5'] = 0
            if '6' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['6'] = 0
            if '7' not in volume_model[time_window_start][tollgate_id][direction]:
                volume_model[time_window_start][tollgate_id][direction]['7'] = 0
            if '0\n' not in volume_type[time_window_start][tollgate_id][direction]:
                volume_type[time_window_start][tollgate_id][direction]['0\n'] = 0
            if '1\n' not in volume_type[time_window_start][tollgate_id][direction]:
                volume_type[time_window_start][tollgate_id][direction]['1\n'] = 0
            if '\n' not in volume_type[time_window_start][tollgate_id][direction]:
                volume_type[time_window_start][tollgate_id][direction]['\n'] = 0     
            if '0' not in volume_has_etc[time_window_start][tollgate_id][direction]:
                volume_has_etc[time_window_start][tollgate_id][direction]['0'] = 0
            if '1' not in volume_has_etc[time_window_start][tollgate_id][direction]:
                volume_has_etc[time_window_start][tollgate_id][direction]['1'] = 0
            out_line = ','.join(['"' + str(tollgate_id) + '"', 
                             '"[' + str(time_window_start) + ',' + str(time_window_end) + ')"',\
                             '"' + str(direction) + '"',\
                             '"' + str(volumes[time_window_start][tollgate_id][direction]) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['0']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['1']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['2']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['3']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['4']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['5']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['6']) + '"',\
                             '"' + str(volume_model[time_window_start][tollgate_id][direction]['7']) + '"',\
                             '"' + str(volume_type[time_window_start][tollgate_id][direction]['0\n']) + '"',\
                             '"' + str(volume_type[time_window_start][tollgate_id][direction]['1\n']) + '"',\
                             '"' + str(volume_has_etc[time_window_start][tollgate_id][direction]['1']) + '"'\
                           ]) + '\n'
            fw.writelines(out_line)
fw.close()   
```
